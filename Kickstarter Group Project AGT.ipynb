```python
import pandas as pd
import numpy as np

df = pd.read_excel('ks-projects-201612.xlsx')
df.head()
```

```python
#keep only failed and successful
df = df.drop(df[df['state'] == 'canceled'].index)
df = df.drop(df[df['state'] == 'live'].index)
df = df.drop(df[df['state'] == 'undefined'].index)
df = df.drop(df[df['state'] == 'suspended'].index)

df = df.drop(df[df['duration']>60].index)
#projects longer than 60 days are no longer possible

df.drop('ID', axis=1, inplace=True)

df.head()
```

```python
df.loc[df['state']=='failed','funded']=0
df.loc[df['state']=='successful','funded']=1
```

```python
df.loc[df['category']!='Playing Cards','Playing Cards']=0
df.loc[df['category']=='Playing Cards','Playing Cards']=1

df.loc[df['main_category']!='Games','Games']=0
df.loc[df['main_category']=='Games','Games']=1
```

```python
df.loc[df['country']!='US', 'USA']=0
df.loc[df['country']=='US', 'USA']=1

df.loc[df['currency']!='USD', 'currencyType']=0
df.loc[df['currency']=='USD', 'currencyType']=1
```

```python
df.loc[df['pledged']<df['goal'], 'goalSuccess']=0
df.loc[df['pledged']>=df['goal'], 'goalSuccess']=1
```

```python
def duration(x):
    if x<=30:
        return 0
    if 30<x<=60:
        return 1
    
df['durationType'] = df['duration'].apply(duration)
```

```python
def backers(x):
    if x < 50:
        return 0
    if 50 < x <=100:
        return 1
    if 100 < x <=500:
        return 2
    if 500 < x <=1000:
        return 3
    if 1000 < x <=5000:
        return 4
    if 5000 < x <=10000:
        return 5
    if 10000 < x <=15000:
        return 6
    if 15000 < x <=20000:
        return 7
    if 20000 < x <=25000:
        return 8
    if x > 25000:
        return 9
```

```python
df['backersType'] = df['backers'].apply(backers)
```

```python
df.loc[df['backers']<=1000, 'backersType1000'] = 0
df.loc[df['backers']>1000, 'backersType1000'] = 1
```

```python
df['avg_donation'] = df['pledged']/df['backers']
```

```python
df.head()
```

```python
df2 = df.filter(['funded','Games','currencyType','avg_donation','USA','goalSuccess','duration','durationType','backers','backersType', 'backersType1000','Playing Cards',], axis=1)
df2.head()
```

```python
df2.corr()
```

```python
#maybe try to narrow down more categories

#look for relationships and then perform regression

#backersType related
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='funded ~ backersType + USA + avg_donation + duration -1', data=df2).fit()
result.summary()
```
